```python
import numpy as np
import json 
```

```python
np.random.seed(42)
total_points=1000
```

```python
x=np.random.uniform(-200,200,total_points)
y=np.random.uniform(-400,400,total_points)
z=np.random.uniform(-100,200,total_points) #1d array created
print(x[:3])
```

```python
points=[]#x,y,z values
for c in range(total_points):
    points.append({"x":float(x[c]),"Y":float(y[c]),"Z":float(z[c])})
print(points[:4])
    
```

```python
with open("data.json","w")as file:
    data=json.dump(points,file,indent=4)
print(f"{total_points} data generated successfully")
```
